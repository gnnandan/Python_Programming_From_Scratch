## Important String Methods

1. capitalize() - Converts the first character to upper case
2. casefold() - Converts string into lower case
3. center(length, character -> optional) - Returns a centered string
4. string.count(value, start -> optional, end -> optional) - Returns the number of times a specified value occurs in a string
5. encode(encoding=encoding -> optional, errors=error -> optional) - Returns an encoded version of the string
6. endswith() - Returns true if the string ends with the specified value
7. find(value, start -> optional, end -> optional) - Searches the string for a specified value and returns the position of where it was found
8. format(value1, value2...) - Formats specified values in a string
9. format_map() - Formats specified values in a string
10. index() - Searches the string for a specified value and returns the position of where it was found
11. isalnum() - Returns True if all characters in the string are alphanumeric
12. isalpha() - Returns True if all characters in the string are in the alphabet
13. isascii() - Returns True if all the characters are ascii characters (a-z)
14. isdecimal() - Returns True if all characters in the string are decimals
15. isdigit() - Returns True if all characters in the string are digits
16. isidentifier() - Returns True if the string is an identifier
17. islower() - Returns True if all characters in the string are lower case
18. isnumeric() - Returns True if all characters in the string are numeric
19. isprintable() - Returns True if all characters in the string are printable
20. isspace() - Returns True if all characters in the string are whitespaces
21. istitle() - Returns True if the string follows the rules of a title
22. isupper() - Returns True if all characters in the string are upper case
23. join(iterable) - Joins the elements of an iterable to the end of the string
24. ljust(length, character) - Returns a left justified version of the string
25. rjust(length, character) - Returns a right justified version of the string
26. lower() - Converts a string into lower case
27. upper() - Converts a string into upper case
28. strip(characters -> optional) - Returns a trimmed version of the string
29. lstrip(characters -> optional) - Returns a left trim version of the string
30. rstrip(characters -> optional) - Returns a right trim version of the string
31. maketrans(x, y -> optional, z -> optional) - Returns a translation table to be used in translations
32. partition(value) - Returns a tuple where the string is parted into three parts
33. rpartition(value) - Returns a tuple where the string is parted into three parts
34. removeprefix(string) -  removes the prefix and returns the rest of the string.
35. removesuffix(string) -  removes the prefix and returns the rest of the string.
36. replace(oldvalue, newvalue, count -> optional) - Returns a string where a specified value is replaced with a specified value
37. rfind(value, start -> optional, end -> optional) - Searches the string for a specified value and returns the last position of where it was found
38. rindex(value, start -> optional, end -> optional) - Searches the string for a specified value and returns the last position of where it was found
39. split(separator -> optional, maxsplit -> optional) - Splits the string at the specified separator, and returns a list
40. swapcase() - Swaps cases, lower case becomes upper case and vice versa
41. title() - Converts the first character of each word to upper case
42. translate(table) - Returns a translated string
43. zfill(len) - Fills the string with a specified number of 0 values at the beginning

### RegEix expression
- This RegEx module provides several functions for working with regular expressions, including <code>search, split, findall,</code> and <code>sub</code>. 


```python
# pattern matching
import re as regex

s1 = "We grow by growing others through Curious Developers Community"

pattern = r"grow"

search = regex.search(pattern, s1)

if(search):
  print("Pattern matched")
else:
  print("Pattern not matched")
```

```python
# grouping special symbols and digits
import re as regex
text = "My Phone number is 9353493539 %^&*890"

patternDigit = r"\d\d\d\d\d\d\d\d\d\d"
patternSpecialSymbol = r"\W"

# finding individual pattern
digitMatch = regex.search(patternDigit, text)

# findding all pattern at once
symbolsMatch = regex.findall(patternSpecialSymbol, text)
digitMatchFindall = regex.findall(patternDigit, text)

if(digitMatch and symbolsMatch and digitMatchFindall):
  print("Curious Team Number: ",digitMatch.group())
  print("Special Symbol: ",symbolsMatch)
  print("Digits: ",digitMatchFindall)
else:
  print("match not found")
```
