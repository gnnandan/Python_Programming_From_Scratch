```python
# program that dynamically take which mathematica operation is to do using Class & Objects

class calculator:
  def __init__(self, a, b):
    self.a = a
    self.b = b
  
  def add(self):
    return self.a + self.b
  
  def sub(self):
    return self.a - self.b
  
  def prod(self):
    return self.a * self.b
  
  def div(self):
    return self.a / self.b
  
  def mod(self):
    return  self.a % self.b
  

a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

# object creation
obj = calculator(a,b)

choice = 1
while(choice!=0):
  print("""
        Select the choice
        ====================
        0. Exit
        1. Addition
        2. Subtraction
        3. Multiplication
        4. Division
        5. Modulus
        """)
  choice = int(input("Enter Your Choice or '0' to exit: "))
  def switch(choice):
    if(choice==1):
      print("Result is: ",obj.add())
    elif(choice==2):
      print("Result is: ",obj.sub())
    elif(choice==3):
      print("Result is: ",obj.div())
    elif(choice==4):
      print("Result is: ",obj.prod())
    elif(choice==5):
      print("Result is: ",obj.mod())
    elif(choice==0):
      print("Exiting...")
    else:
      print("Invalid Choice")
  
  
  switch(choice)
```

```python
# Python program to append, delete, display of list elements 

class listOperation:
  def __init__(self):
    self.num = []
  def add(self,a):
    return self.num.append(a)
  def delete(self,b):
    return self.num.remove(b)
  def display(self):
    return self.num

obj = listOperation()

choice =1
while(choice !=0):
  print("0: Exit")
  print("1: append")
  print("2: Delete")
  print("3: Display")
  choice = int(input("Enter your choice: "))
  if(choice == 1):
    num = int(input("Enter the elements to append to list: "))
    obj.add(num)
    print("After appending elements the final list values: ", obj.display())
    
  elif(choice == 2):
    element = int(input("Enter the elements to remove from the list: "))
    obj.delete(element)
    print("After deleting element from the list: ", obj.display())
  elif(choice == 3):
    print("The list elements: ", obj.display())
  elif(choice == 0):
    print("Exiting....")
    print("The final list elements: ", obj.display())
    break
  else:
    print(f"{choice} is not a valid choice")
    
print()
```
