```python
# creating dictionary
myDict = {1:"One",2:"Two",3:"Three",4:"Four",5:"Five"}
print(myDict)

# checking datatype
print(type(myDict))

# accessing elements
print(myDict[2])

# accessing element with get property
print(myDict.get(2))

# accessing all the keys
print(myDict.keys())

# accessing all the values
print(myDict.values())

# accessing all the Keys, Values
print(myDict.items())

# updating the element

myDict[2] = "TWO"
print("After updating the element: {}".format(myDict))
myDict.update({3:"THREE"})
print("After updating using update method the element: {}".format(myDict))

# removing the element
myDict.pop(3)
print("After removing the element: {}".format(myDict))

# iterating over all elements
print("Iterating over keys")
for key in myDict.keys():
  print("Keys are {} ".format(key))

print("Iterating over values")
for values in myDict.values():
  print("Values are {} ".format(values))

print("Iterating over all elements")
for key,values in myDict.items():
  print("Keys are {} and Values are {}".format(key,values))
```

## Important Methods
1. clear() - Removes all the elements from the dictionary
2. copy() - Returns a copy of the dictionary
3. fromkeys(keys, value -> optional) - Returns a dictionary with the specified keys and value
4. get(keyname, value -> optional) - Returns the value of the specified key
5. items() - Returns a list containing a tuple for each key value pair
6. keys() - Returns a list containing the dictionary's keys
7. pop(keyname, defaultvalue -> optional) - Removes the element with the specified key
8. popitem() - Removes the last inserted key-value pair
9. setdefault(keyname, value -> optional) - Returns the value of the specified key. If the key does not
10. exist: insert the key, with the specified value
11. update() - Updates the dictionary with the specified key-value pairs
12. values() - Returns a list of all the values in the dictionary

```python
myDict = {1:"One",2:"Two",3:"Three",4:"Four",5:"Five"}

# get()
print(myDict.get(3))

# update()
newDict = ["one","two","three"]
myDict.update({2:newDict})
print("After updating the dictionary: {}".format(myDict))
myDict.update({1:"ONE"})
print("After updating the dictionary: {}".format(myDict))

# keys()
print(myDict.keys())

# values()
print(myDict.values())

# items()
print(myDict.items())

# fromkeys()
dict1 = {1:"One",2:"Two",3:"Three",4:"Four",5:"Five"}
dict2 = {1:"aaa",2:"bbb",3:"ccc",4:"ddd",5:"eee"}
dict1.fromkeys(dict1,dict2)

# copy
copyDict = myDict.copy()
print("Copied dictionary: {}".format(copyDict))

# pop()
myDict.pop(2,"notfound")
print("After popping element {}".format(myDict))

# popitem()
myDict.popitem()
print("After popping element {}".format(myDict))

myDict.clear()
print("After clearing dictionary {}".format(myDict))
```

## Looping in Dictionaries


```python
myDict = {1:"One",2:"Two",3:"Three",4:"Four",5:"Five"}

# looping over keys
for keys in myDict.keys():
  print(keys)

# looping over values
for values in myDict.values():
  print(values)

# looping over keys and values
for keys,values in myDict.items():
  print(f"key's data {keys} value's data {values}")

"""
1
2
3
4
5
One
Two
Three
Four
Five
key's data 1 value's data One
key's data 2 value's data Two
key's data 3 value's data Three
key's data 4 value's data Four
key's data 5 value's data Five
"""
```

```python
myfamily = {
  "child1" : {
    "name" : "Emil",
    "year" : 2004
  },
  "child2" : {
    "name" : "Tobias",
    "year" : 2007
  },
  "child3" : {
    "name" : "Linus",
    "year" : 2011
  }
}

# accessing keys in Nested Dictionary
print("Keys in Nested Dictionary")
for data in myfamily.keys():
  print(data)
  
# accessing values in Nested Dictionary
print("Values in Nested Dictionary")
for values in myfamily.values():
  print(values)

print("Keys & Values in Nested Dictionary")
for keys,values in myfamily.items():
  print(keys," - ",values)
  
print("Accessing individual element - {}".format(myfamily["child1"]["name"]))
print("Accessing individual element - {}".format(myfamily["child2"]["year"]))
```
